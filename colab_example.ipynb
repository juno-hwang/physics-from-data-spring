```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("tracker_data.csv") 
df1
```

```python
tracker_data = np.array(df1)
t1 = tracker_data[:,0]
y1 = tracker_data[:,1] 
v1 = (y1[1:] - y1[0:-1])/(t1[1:] - t1[0:-1])
a1 = (v1[1:] - v1[0:-1])/(t1[2:] - t1[1:-1])

y1 = (y1 - np.mean(y1))/max(y1)
v1 = (v1 - np.mean(v1))/max(v1)
a1 = (a1 - np.mean(a1))/max(a1)

#plt.scatter(t1[:], y1)
#plt.scatter(t1[1:], v1)
plt.scatter(t1[2:], a1)


#plt.plot(t1[:],np.sin(2.*np.pi/1.2*t1[:]+np.pi))
#plt.plot(t1[1:],np.cos(2.*np.pi/1.2*t1[1:]+np.pi))
plt.plot(t1[1:],-np.sin(2.*np.pi/1.2*t1[1:]+np.pi))
plt.xlim(0,3)
plt.show()
```
